```python
#Q1
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt  

df = pd.read_csv(R"C:\Users\dell\OneDrive\Desktop\cleaned.csv")
df.rename(columns={"amount":"sales"}, inplace=True)

best_year = df.groupby('year')['sales'].sum().idxmax()
print("Best Year of Sales:", best_year)
```

```python
#Q2
best_month = df.groupby('month')['sales'].sum().idxmax()
print("Best Month of Sales:", best_month)
```

```python
#Q3
print(df.groupby('year')['sales'].sum())
```

```python
#Q4
top_brand = df[(df['year']>=2015) & (df['year']<=2018)] \
              .groupby('brand')['sales'].sum().idxmax()
print("Top Brand:", top_brand)
```

```python
#Q5
top_brand = df[df['year']==2016].groupby('brand')['sales'].sum().idxmax()
df[(df['year']==2016) & (df['brand']==top_brand)] \
  .groupby('month')['sales'].sum().plot(kind='bar' , color='cyan')

plt.title("Top Brand Sales in 2016")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()
```

```python
#Q6
result = df.groupby(['category','item_id'])['sales'].sum()
print(result.groupby(level=0).idxmax())
```

```python
#Q7
least_product = df[(df['year']>=2015) & (df['year']<=2018)] \
                  .groupby(['brand','item_id'])['sales'].sum().idxmin()
print("Least Sold Product:", least_product)
```

```python
#8
df['rating'].plot(kind='hist')
plt.title("Distribution of Ratings")
plt.xlabel("Rating")
plt.ylabel("Frequency")
plt.show()
```

```python
#Q9
top_rated = df[(df['year']>=2015) & (df['year']<=2018)] \
              .groupby('brand')['rating'].count().idxmax()
print("Most Rated Brand:", top_rated)
```

```python
#Q10
df.groupby('category')['sales'].sum().plot(kind='pie', autopct='%1.1f%%')
plt.title("Category Wise Sales Percentage")
plt.ylabel("")
plt.show()
```

```python
#Q11
brand_group = df.groupby('brand')['sales'].sum().nlargest(8)
plt.figure()
plt.pie(brand_group.values,
        labels=brand_group.index,
        autopct='%0.2f%%')
plt.title("Top Brands with Highest Sales Percentage")
plt.show()
```

```python
#Q12
df['gender'].value_counts().plot(kind='bar', color='plum')
plt.title("Gender Wise Customer Distribution")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

```python
#Q13
df.groupby(['gender','category'])['quantity'].sum().unstack().plot(kind='bar')
plt.title("Quantity Sold by Gender and Category")
plt.xlabel("Gender")
plt.ylabel("Quantity")
plt.show()
```
